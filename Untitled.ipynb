```python
import pandas as pd
import glob

# just defining some helpful variables
img_type_lst = ['small', 'big', 'mixture']
thread_list = [2, 4, 6, 8, 12]

## CALCULATE AMDAHLS
colnames = [('parFile', img_type) for img_type in img_type_lst]
colnames.extend([('parSlices', img_type) for img_type in img_type_lst])

## amdahls dataframe
amdahl = pd.DataFrame(columns = colnames)
amdahl.index.name = 'Thread Count'

## best parfile/parslice sequential time
par_img_dict = dict()
for img_type in ['small', 'big', 'mixture']:
    # find best runtime
    mintime = 100000
    for file in glob.glob(f'data/sequential_{img_type}_*.txt'):
        time_df = pd.read_csv(file, sep = '\t')
        ov_time = time_df.loc[0, 'Overall']
        if ov_time < mintime:
            bestdf = time_df.copy()
            mintime = ov_time
    # find parallel proportion
    ov_time = bestdf.loc[0, 'Overall']
    parfiles_time = bestdf.loc[0, 'Parfiles']
    parslices_time = bestdf.loc[0, 'Parslices']

    # add to dict
    par_img_dict[f'parfiles_{img_type}'] = [ov_time, parfiles_time]
    par_img_dict[f'parslices_{img_type}'] = [ov_time, parslices_time]
    # fraction of time
    parfile_fraction = parfiles_time/ov_time
    parslices_fraction = parslices_time/ov_time

    # calculate amdahl and add to table
    parfile_val = []
    parslice_val = []
    for threads in [2, 4, 6, 8, 12]:
        parfile_val.append(1/((1-parfile_fraction) + parfile_fraction/threads))
        parslice_val.append(1/((1-parslices_fraction) + parslices_fraction/threads))
    amdahl[('parFile', img_type)] = parfile_val
    amdahl[('parSlices', img_type)] = parslice_val
# fix index
amdahl.index = [2, 4, 6, 8, 12]

amdahl.to_csv('analysis/amdahl.csv')
```

```python
## CREATE SPEEDUP GRAPHS
for partype in ['parfiles', 'parslices']:
    small_lst = []
    big_lst = []
    mixture_lst = []
    for thread in thread_list:
        small_lst.append(par_img_dict[f'{partype}_small'][1]/float(pd.read_csv("data/"+f"{partype}_{thread}_small.txt").columns[0]))
        big_lst.append(par_img_dict[f'{partype}_big'][1]/float(pd.read_csv("data/"+f"{partype}_{thread}_big.txt").columns[0]))
        mixture_lst.append(par_img_dict[f'{partype}_mixture'][1]/float(pd.read_csv("data/"+f"{partype}_{thread}_mixture.txt").columns[0]))

    
    fig, ax = plt.subplots()
    ax.plot(thread_list, small_lst, label ='Small')
    ax.plot(thread_list, big_lst, label ='Big')
    ax.plot(thread_list, mixture_lst, label ='Mixture')

    ax.set_xlabel("Thread Count")
    ax.set_ylabel("Speedup")
    ax.legend()
    ax.set_title(f'Speedup for {partype} parallelization')
    plt.show()
    fig.savefig(f'speedup-{partype}.png')
```
